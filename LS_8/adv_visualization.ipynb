```python
# import library
from wordcloud import WordCloud, STOPWORDS
```

```python
# open file
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
# iniasi wordcloud
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
)

# generate wordcloud
alice_wc.generate(alice_novel)
```

```python
stopwords = set(STOPWORDS)
stopwords
```

```python
# tampilkan wordcloud
import matplotlib.pyplot as plt
plt.imshow(alice_wc, interpolation='bilinear');
plt.axis('off');
```

```python
# metode fig
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(24)

plt.imshow(alice_wc, interpolation='bilinear');
plt.axis('off');
```

```python
stopwords.add('said')
# iniasi wordcloud
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate wordcloud
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(24)


plt.imshow(alice_wc, interpolation='bilinear');
plt.axis('off');
```

```python
# masking
import numpy as np
from PIL import Image

alice_mask = np.array(Image.open("alice_mask.png"))
```

```python
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(24)


plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear');
plt.axis('off');
```

```python
# iniasi wordcloud
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords,
    mask=alice_mask
)

# generate wordcloud
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(24)


plt.imshow(alice_wc, interpolation='bilinear');
plt.axis('off');
```

### Regression Plots

```python
import pandas as pd
df_can = pd.read_excel("Canada.xlsx",
                        sheet_name='Canada by Citizenship',
                        skiprows=20,
                        skipfooter=2)
df_can.head()
```

```python
# bersihkan data
# hapus kolom yang tidak diperlukan
kol = ['AREA', 'REG', 'DEV', 'Type', 'Coverage']
df_can.drop(kol, axis=1, inplace=True)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country',
                        'AreaName':'Continent',
                        'RegName':'Region'}, inplace=True)
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
# import seaborn
import seaborn as sns
```

```python
years = np.arange(1980, 2014)
df_total = pd.DataFrame(df_can[years].sum())
df_total
```

```python
years = np.arange(1980, 2014)
df_total = pd.DataFrame(df_can[years].sum())
df_total.index = map(float, df_total.index)
df_total.reset_index(inplace=True)
df_total.columns = ['years', 'total']
df_total.head()
```

```python
sns.regplot(x='years', y='total', data=df_total, color='purple', marker='*');
```

```python
# open data
url = "https://github.com/ardhiraka/PFDS_sources/raw/master/Police_Department_Incidents_-_Previous_Year__2016_.csv"
df_incidents = pd.read_csv(url)
df_incidents.head()
```

```python
df_incidents.shape
```

```python
# limit data hanya 100 teratas
limit = 100
df_incidents = df_incidents.iloc[:limit, :]
df_incidents.head()
```

```python
df_incidents.shape
```

```python
latitude = 37.77
longitude = -122.43
```

```python
# generate san fransisco map
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

sanfran_map
```

```python
# inisiasi fitur grup
incidents = folium.map.FeatureGroup()

# looping ke dataset untuk long dan lat nya
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6,
        )
    )

sanfran_map.add_child(incidents)
```

```python
# inisiasi fitur grup
incidents = folium.map.FeatureGroup()

# looping ke dataset untuk long dan lat nya
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6,
        )
    )

latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)

sanfran_map.add_child(incidents)
```

```python
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Descript):
    folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            popup=label,
            fill_color='blue',
            fill_opacity=0.6,
    ).add_to(sanfran_map)

sanfran_map
```

```python
from folium import plugins

sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

incidents = plugins.MarkerCluster().add_to(sanfran_map)

for lat, lng, label in zip(df_incidents.Y, df_incidents.X, df_incidents.Descript):
    folium.CircleMarker(
            [lat, lng],
            popup=label,
            icon=None
    ).add_to(incidents)

sanfran_map
```
